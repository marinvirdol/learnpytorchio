### Data (preparing and loading)
Data can be almost anything.
* images
* videos
* audio
* text

Machine Learning is a game of two parts:
1. **Turn your data**, whatever it is, into **numerical representation**
2. **Pick or Build** a model to **learn the representation** as best as possible.

To showcase this, let's create some *known* data using the linear regression formula.

We'll use a linear regression formula to create our data as a straight line with  ***known parameters***. Then, we'll use PyTorch to see if we can build a model to estimate these parameters using ***gradient descent***.

A parameter is something that the model learns.

````python
# create *known* parameters
weight = 0.7 # whatever we want
bias = 0.3

start = 0
end = 1
step = 0.02

X = torch.arange(start, end, step).unsqueeze(dim=1)

y = weight * X + bias # we know what the relationship between X and y is

X[:10], y[:10]
````

````python
len(X), len(y)
````

We are now going to build a model that can learn the relationship between `X` (features) and `y` (labels).

In this particular example we already know what the relationship between X and y is: ```y = weight * X + bias```. However, in reality we will not know this relationship.
